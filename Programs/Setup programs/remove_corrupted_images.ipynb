```python
import os
import cv2
import imghdr
from PIL import Image
import shutil
```

```python
def check_images(s_dir, ext_list):
    bad_images = []
    s_list = os.listdir(s_dir)
    for klass in s_list:
        klass_path = os.path.join(s_dir, klass)
        print("processing class directory ", klass)
        if os.path.isdir(klass_path):
            file_list = os.listdir(klass_path)
            for f in file_list:
                f_path = os.path.join(klass_path, f)
                tip = imghdr.what(f_path)
                if ext_list.count(tip) == 0:
                    bad_images.append(f_path)
                if os.path.isfile(f_path):
                    try:
                        img = cv2.imread(f_path)
                        shape = img.shape
                    except:
                        bad_images.append(f_path)
                else:
                    print(
                        "*** fatal error, you a sub directory ",
                        f,
                        " in class directory ",
                        klass,
                    )
        else:
            print(
                "*** WARNING*** you have files in ",
                s_dir,
                " it should only contain sub directories",
            )
    return bad_images
```

```python
source_dir = r"C:\Users\zuraj\Downloads\MiniProjectNew\Dataset"
acceptable_exts = ["jpg", "png", "jpeg"]

# Isolating all unacceptable file types
bad_file_list = check_images(source_dir, acceptable_exts)

if len(bad_file_list) != 0:
    print("improper image files are listed below")
    for i in range(len(bad_file_list)):
        print(bad_file_list[i])
else:
    print(" no improper image files were found")
```

```python
def get_image_format(file_path):
    image_format = imghdr.what(file_path)
    return image_format
```

```python
def convert_to_jpeg(image_path, output_path):
    image = Image.open(image_path)
    if image.mode != "RGB":
        image = image.convert("RGB")
    image.save(output_path, format="JPEG")
    print(f"Converted file: {output_path}")
```

```python
convert_acceptable_exts = ["bmp", "tiff", "gif", "jpeg", "png"]

for i in range(len(bad_file_list)):
    file_path = bad_file_list[i]
    if get_image_format(file_path) in convert_acceptable_exts:
        output_path = file_path.split(".")[0] + "converted.jpg"
        convert_to_jpeg(file_path, output_path)
        shutil.move(output_path, file_path)
        print("Replaced file: ", file_path)
    else:
        os.remove(file_path)
        print("Removed file:", file_path)
```
