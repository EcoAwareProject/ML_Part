```python
import os
import hashlib
from PIL import Image
```

```python
def get_image_hash(image_path):
    try:
        image = Image.open(image_path)
        image_hash = hashlib.md5(image.tobytes()).hexdigest()
        return image_hash
    except Exception as e:
        print(f"Error processing image: {image_path}")
        return None
```

```python
def remove_duplicate_images(folder_path):
    image_hashes = {}
    duplicate_images = []

    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        if os.path.isfile(file_path):
            image_hash = get_image_hash(file_path)
            if image_hash is not None:
                if image_hash not in image_hashes:
                    image_hashes[image_hash] = file_path
                else:
                    duplicate_images.append(file_path)
            else:
                print(f"Skipped file: {file_path}")

    for duplicate_image in duplicate_images:
        os.remove(duplicate_image)
        print(f"Removed duplicate image: {duplicate_image}")

    print(f"Total duplicate images removed: {len(duplicate_images)}")


dir_path = r"C:\Users\zuraj\Downloads\MiniProjectNew\Dataset"
for folder in os.listdir(dir_path):
    print("Folder name:", folder)
    folder_path = os.path.join(dir_path, folder)
    remove_duplicate_images(folder_path)
```
